```python
import datetime

#Def menu awal
def show_menu():
    print ("Silahkan pilih bantuan yang bisa aku berikan:")
    print ("1. Menghitung dengan kalkulator")
    print ("2. Mencatat jadwal agenda harianmu")
    print ("3. Tentang kamu dan hari ini")
    print ("4. Cerita dan curhatmu")
    print ("5. Keluar")

#Def kalkulator
def tambah(x, y):
    return x + y
 
def kurang(x, y):
    return x - y
 
def kali(x, y):
    return x * y
 
def bagi(x, y):
    return x / y

def show_kalkulator():
    print("Menu hitungan:")
    print("1. Penjumlahan")
    print("2. Pengurangan")
    print("3. Perkalian")
    print("4. Pembagian")
    print("5. Keluar")

#Def agenda
def show_menu_3():
    print("Apa agendamu?")
    print("1. Daftar agenda")
    print("2. Menambah agenda")
    print("3. Keluar")

def tampilkan_agenda():
    print("Daftar agenda:")
    for agenda in daftar_agenda:
        print("Rencana kegiatanmu: " + agenda["Kegiatan"])
        print("Dilakukan tanggal: " + agenda["Tanggal"])
        print("Dilakukan jam: " + agenda["Jam"])

def tambah_agenda():
    Kegiatan = input("Apa rencana kegiatanmu: ")
    Tanggal = input("Tanggal berapa itu dilakukan (format DD-MM-YYYY): ")
    Jam = input("Dilakukan jam berapa (format HH:MM): ")
    agenda = {
        "Kegiatan": Kegiatan,
        "Tanggal": Tanggal,
        "Jam": Jam,
    }
    daftar_agenda.append(agenda)
    print(" ")
    print("Oke, agendamu berhasil ditambahkan")

def namafile():
    f=open("Agendamu.txt", "w+")
    f.write(str(daftar_agenda))
    f.close()

daftar_agenda = []   

#Def tentang kamu dan hari ini
def show_menu_2():
    print ("Silahkan pilih pertanyaan kamu:")
    print ("1. Nama kamu siapa?")
    print ("2. Kamu siapa?")
    print ("3. Tanggal berapa hari ini?")
    print ("4. Hari ini hari apa?")
    print ("5. Saya tidak mau bertanya lagi")

#Parameter tentang kamu dan hari ini
x = datetime.datetime.now()
tanggal = x.strftime("%d-%m-%Y")
hari = x.strftime("%A")
hari_ind = {"Sunday": "Minggu","Monday": "Senin","Tuesday": "Selasa","Wednesday": "Rabu","Thursday": "Kamis",
            "Friday": "Jumat","Saturday": "Sabtu"
}
      
#Pengantar
print("Hai aku AIs, asisten virtual pertamamu")
print("Siapa namamu?")

nama  = str(input("Masukkan nama: " ))

default = "Maaf, aku belum paham yang kamu tanyakan"

print(" ")

print("Salam kenal {}".format(nama))
print("Apakah ada yang bisa aku bantu?")

    # list jawaban untuk pertanyaan tentang nama
jawaban_nama = [
      "Nama saya  {}".format(nama),
      "Orang-orang memanggil saya {}".format(nama),
      "Panggil saja saya {}".format(nama)
    ]

    # list jawaban untuk pertanyaan tentang tanggal
jawaban_tanggal = [
      "Hari ini tanggal {}".format(tanggal),
      "Ya ampun masa tidak tahu, hari ini kan hari {}".format(hari_ind[hari])
    ]

#Menu awal
while True:
    print(" ")
    show_menu()
    menu_awal = input("Pilih hanya angka aja ya:")
    print(" ")

    if menu_awal == "1":
        #Kalkulator
        while True:
            print(" ")
            show_kalkulator()
            operasi = input("Ayo pilih menu (hanya tulis angka aja ya):")
            print(" ")
               
            x = int(input("Masukan angka pertama (tolong isi angka sembarang jika kamu telah memilih menu keluar): "))
            y = int(input("Masukan angka ke dua (tolong isi angka sembarang jika kamu telah memilih menu keluar): "))
            
            if operasi == "1":
                print(x , "+" , y , "=" , tambah(x, y))
            elif operasi == "2":
                print(x , "-", y, "=", kurang(x, y))
            elif operasi == "3":
                print(x, "x", y, "=", kali(x, y))
            elif operasi == "4":
                print(x, ":", y, "=", bagi(x, y))
            elif operasi == "5":
                print("Oke, terima kasih sudah menggunakan kalkulator ini")
            else:
                print(default)
            break
            
    elif menu_awal == "2":
        #Agenda
        while True:
            print(" ")
            show_menu_3()
            menu_agenda = input("Pilih hanya angka aja ya:")
            print(" ")

            if menu_agenda == "1":
                tampilkan_agenda()
            elif menu_agenda == "2":
                tambah_agenda()
            elif menu_agenda == "3":
                print("Yasuda kalo gitu")
                break
            else:
                print(default)

        namafile()
        
    elif menu_awal == "3":
        #Tentang kamu dan hari ini
        while True:
            print(" ")
            show_menu_2()
            menu_kamu = input("Pilih hanya angka aja ya:")
            print(" ")

            if menu_kamu == "1": 
                print(jawaban_nama[0] +","+" "+ jawaban_nama[1])
            elif menu_kamu == "2": 
                print(jawaban_nama[2])
            elif menu_kamu == "3": 
                print(jawaban_tanggal[0])
            elif menu_kamu == "4": 
                print(jawaban_tanggal[1])
            elif menu_kamu == "5": 
                print("Yasuda kalo gitu")
                break
            else: 
                print (default)  
                
    elif menu_awal == "4":
        while True:
            Cerita = input('Coba ceritakan lebih banyak lagi')
            print(" ")

            if Cerita.find("bodoh")!= -1 or Cerita.find("gagal")!= -1 or Cerita.find("tidak berhasil")!= -1 or Cerita.find("kalah")!= -1 or Cerita.find("jelek")!= -1 or Cerita.find("tipu")!= -1:
                input('Kenapa kamu berpikir begitu?: ')
            elif Cerita.find("kecewa")!= -1 or Cerita.find("putus asa")!= -1 or Cerita.find("demotivasi")!= -1 or Cerita.find("harap")!= -1 or Cerita.find("sedih")!= -1:
                input ('Sudah berapa lama kamu merasa seperti ini?: ')
            elif Cerita.find("berhasil")!= -1 or Cerita.find("menang")!= -1 or Cerita.find("termotivasi")!= -1 or Cerita.find("hadiah")!= -1 or Cerita.find("senang")!= -1:
                input ('Itu sangat menarik!: ')
            else:
                print('Oh WOW!' + ':)')
                break
                
    elif menu_awal == "5":  
        print("Oke, sampai jumpa")
        break
    else:
        print(default)
```
